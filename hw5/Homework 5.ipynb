```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import surprise
from surprise.model_selection import cross_validate
from surprise.prediction_algorithms.matrix_factorization import SVD
from surprise.prediction_algorithms.knns import KNNBasic
```

```python
ratings_df = pd.read_csv("ratings.csv")
```

```python
MAX_K = 150
```

```python
def pmf_evaluation():
    print("Beginning PMF evaluation")
    
    # Convert data to a surprise dataset so that surprise library functions work with it
    data = surprise.Dataset.load_from_df(ratings_df[['userId', 'movieId', 'rating']], 
                                         surprise.Reader(rating_scale=(1, 5)))
    
    # SVD with no bias is identical to PMF (according to the surprise documentation)
    PMF = SVD(biased = False, verbose=True)

    # Perform 5-fold cross validation with RSME and MAE
    results = cross_validate(PMF, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
    
    # Convert the cross-validation results to a pandas dataframe to leverage built-in csv-ification
    results_df = pd.DataFrame(results)
    results_df.to_csv("pmf_evaluation.csv")
```

```python
def cf_evaluation(name: str, user_based: bool, k: int):
    print(f"Beginning {'user' if user_based else 'item'}-based CF evaluation with {name} similarity and k={k}")
    
    # Convert data to a surprise dataset so that surprise library functions work with it
    data = surprise.Dataset.load_from_df(ratings_df[['userId', 'movieId', 'rating']], 
                                         surprise.Reader(rating_scale=(1, 5)))
    
    sim_options = {
        'name': name,            # cosine, msd, or pearson
        'user_based': user_based # true for user cf, false for item cf
    }
    
    # KNNBasic is a "basic collaborative filtering algorithm" (according to the surprise documentation)
    CF = KNNBasic(k=k, sim_options=sim_options, verbose=True)
    
    
    print("Beginning the slow part")
    # Perform 5-fold cross validation with RSME and MAE
    results = cross_validate(CF, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
    
    # Convert the cross-validation results to a pandas dataframe to leverage built-in csv-ification
    results_df = pd.DataFrame(results)
    results_df.to_csv(f"cf_evaluation_{'user' if user_based else 'item'}_{name}_{k}.csv")
```

```python
def run_all_evaluations():
    pmf_evaluation()
    
    for k in range(1, MAX_K + 1):
        for metric in ('cosine', 'msd', 'pearson'):
            for user_based in (True, False):
                cf_evaluation(metric, user_based, k)
```

```python
# Loads the CF data from the form in which it was saved by cf_evaluation
# and reorganizes it into a single pandas dataframe designed 
# for use in creating meaningful plots
def reorganize_data():
    data = dict()
    total_time = [0] * MAX_K
    
    for user_based in ('user', 'item'):
        for metric in ('cosine', 'msd', 'pearson'):
            data[f'{user_based}_{metric}_rmse'] = []
            data[f'{user_based}_{metric}_mae'] = []
            
            for k in range(1, MAX_K + 1):
                df = pd.read_csv(f'output/cf_evaluation_{user_based}_{metric}_{k}.csv')
                rmse = df['test_rmse'].mean()
                mae = df['test_mae'].mean()
                
                total_time[k-1] += df['fit_time'].sum() + df['test_time'].sum()
                
                data[f'{user_based}_{metric}_rmse'].append(rmse)
                data[f'{user_based}_{metric}_mae'].append(mae)
    
    # This is an analysis of how long it takes to run the calculations
    # and how long it will take to run more of them
    # it's not essential for the actual operation of the system
    if True:
        x = list(range(1, MAX_K+1))
        print(f'Total Time for {MAX_K} iterations: {round(sum(total_time) / 3600, 2)} hours')
        m, b = np.polyfit(x, total_time, 1) 
        print(f'y = {m:.3f}x + {b:.1f}')
        start, end = MAX_K + 1, MAX_K + 50
        integral = (0.5 * m * end * end + b * end) - (0.5 * m * start * start + b * start)
        integral = integral / 3600
        print(f'Estimated time to calculate from k={start} to k={end}: {integral:.2f} hours')
        plt.plot(x, total_time)
        plt.show()
    
    data_df = pd.DataFrame(data)
    data_df.to_csv(f'reorganized_data.csv')
    return data_df
                
```

```python
# Simple pyplot generator for the reorganized data
# Plots all similarity metrics in one figure for a given user/item and rmse/mae
def plot_data(user_or_item: str, rmse_or_mae: str, df: pd.DataFrame):
    assert user_or_item in ('user', 'item'), "user_or_item must be 'user' or 'item'"
    assert rmse_or_mae in ('rmse', 'mae'), "rmse_or_mae must be 'rmse' or 'mae'"
    
    cosine = df[f'{user_or_item}_cosine_{rmse_or_mae}']
    msd = df[f'{user_or_item}_msd_{rmse_or_mae}']
    pearson = df[f'{user_or_item}_pearson_{rmse_or_mae}']
    
    x = list(range(1, MAX_K + 1))
    
    plt.plot(x, cosine, c='blue', label='cosine')
    plt.plot(x, msd, c='red', label='msd')
    plt.plot(x, pearson, c='green', label='pearson')
    plt.legend()
    
    plt.xlabel('Number of Neighbors')
    plt.ylabel(f'{rmse_or_mae.upper()}')
    
    plt.title(f'{user_or_item.title()}-based CF {rmse_or_mae.upper()}')
    
    plt.savefig(f'plots/plot_{user_or_item}_{rmse_or_mae}.jpg', dpi=1000, bbox_inches='tight')
    plt.show()
```

```python
# Plots a comparison between user-based and item-based, using MSD
def plot_user_vs_item(rmse_or_mae: str, df: pd.DataFrame):
    assert rmse_or_mae in ('rmse', 'mae'), "rmse_or_mae must be 'rmse' or 'mae'"
    
    user = df[f'user_msd_{rmse_or_mae}']
    item = df[f'item_msd_{rmse_or_mae}']
    
    x = list(range(1, MAX_K + 1))
    
    plt.plot(x, user, c='red', label='user')
    plt.plot(x, item, c='blue', label='item')
    plt.legend()
    
    plt.xlabel('Number of Neighbors')
    plt.ylabel(f'{rmse_or_mae.upper()}')
    
    plt.title(f'User-based vs Item-based CF using {rmse_or_mae.upper()} of MSD')
    
    plt.savefig(f'plots/plot_user_vs_item_{rmse_or_mae}.jpg', dpi=1000, bbox_inches='tight')
    plt.show()
```

```python
organized_data = reorganize_data()
```
